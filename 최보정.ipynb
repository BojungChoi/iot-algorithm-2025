2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오

```python
def isQueueFull():
    global SIZE, Queue, front, rear # 전역 변수 사용
    if rear != SIZE - 1 : # rear 가  마지막 인덱스가 아니면 (SIZE -1)
        return False # 큐가 차지않음.
    elif rear == SIZE - 1 and front == -1: # rear가 마지막인덱스까지 차있고 and , front가 처음 위치라면
        return True # 큐가 가득 참
    else: # 큐가 가득 찼지만, front가 비어 있는 공간을 남겨둔 경우 (선형 큐 문제 해결)
        for i in range(front +1, SIZE): # 현재 +1 부분 -> 데이터가 있는 부분
            Queue[i-1] = Queue[i] # 데이터 빈자리에 첫번째 데이터를 옮김
            Queue[i] = None # 데이터를 옮긴 후 None 처리 -> for 문으로 인해 무한반복 -> 한마디로 순차적으로 앞으로 다 땡김

        front -= 1 # 한칸씩 이동
        rear -= 1 # rear도 한칸이동함으로 공간확보
        return False

def isQueueEmpty():
    global SIZE, Queue, front, rear
    if front == rear: # rear가 앞 데이터부터 삭제되어 front와 동일 한 경우
        return True # 리턴
    else:
        return False # Queue에 데이터가있을때.
    
def enQueue(data):
    global SIZE, Queue, front, rear
    if isQueueFull(): # Que가 가득찼을경우 isQueueFull() 함수이동
        print('대기 인원 수 초과') # full 프린터 출력
    else:
        rear += 1 # enQueue 는 rear 를 1 증가
        Queue[rear] = data # 해당위치 data 삽입

def deQueue():
    global SIZE, Queue, front, rear
    if isQueueEmpty(): # 만약 Queue 데이터가 비었으면
        print('대기자 없음') # 프린터를 출력하고
        return None # 종료
    else:
        front += 1 # deQuquq 는 front 를 1 증가 (데이터 자리 확보)
        data = Queue[front] # 데이터 저장
        Queue[front] = None # 저장 후 front data 없애주고
        
        for i in range(front+1, rear+1): # 데이터가 빠지고 나면 앞에 None 이 되기때문에 모든 사람을 한칸씩 앞으로 당긴다
            Queue[i-1] = Queue[i] 
            Queue[i] = None
        front = -1
        rear -= 1        
        
        return data # 끝냄
    
def peek(): # 큐데이터 확인
    global SIZE, Queue, front, rear
    if isQueueEmpty(): # 큐가 비었을때 (isQueueEmpty 함수돌고)
        print('대기자 없음') # 출력
        return None # 끝냄
    else: # 큐가 데이터가 있다면 (front != rear 라면)
        return Queue[front + 1] # front 바로 다음 위치의 데이터 반환 , Front에는 값이없어 1 뒤에 데이터를 본다.(1을 더하는것이아니다.) ,

# 초기화
SIZE = int(input('대기 가능 자리 수 입력 > ')) # 동적으로 input받아 큐 크기를 입력받음
Queue = [None for _ in range(SIZE)] # 입력받은 Size 크기를 None 처리
front = rear = -1 # front와 rear를 초기 상태(-1)로 설정


## 메인 모듈
if __name__ == '__main__' : 
    while True:
        select = input('삽입(I) / 추출(E) / 확인(V) / 종료(Q) --> ').upper() # 전역 변수 선언 할 수있는 Input  동작입력 받기
    
        if select == 'Q': # 종료 선택
            break # if 문 빠져나감
        elif select == 'I': # 삽입 선택
            data = input('대기하시는 분 성함을 입력하세요 > ') # 삽입할 데이터 입력
            enQueue(data) # 입력받은 데이터로 enQueue함수이동
            print(f'현대 대기 인원자 : {Queue}') # 큐 상태 확인
        elif select == 'E': # 데이터 추출
            data = deQueue() # deQueue() 함수 실행 , 출력값 data 변수에 삽입
            if data == None:
                print('대기인원이 없습니다.')
                continue
            print(f' {data} 님 들어가셨습니다.') # data 변수 확인 출력
            print(f'현대 대기 인원자 : {Queue}') # 큐 상태 확인
        elif select == 'V': # 데이터 확인
            data = peek() # peek 함수 확인
            print(f'현대 대기 인원자 : {Queue}')  # 큐 상태 확인

        else:
            print('나머지 동작') # 잘못된 입력 예외 처리
```

3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
데이터의 종류는 [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
import random

# 초기화
memory = [] # 이진탐색할 메모리 공간
root = None # 트리의 루트에 저장할 변수
dataAry = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워', 
           '요아정 요거트컵', '페퍼로니피자주먹밥', '널담 슈톨렌', '딸기마시멜로케이크', 
           '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵']

# 랜덤하게 20개 판매 기록 생성
sellAry = [random.choice(dataAry) for _ in range(20)] # 랜덤하게 20 개의 품목을 선택하여 리스트 저장.

print('오늘 판매된 물건(중복O) -->', sellAry) # 중복값을 포함하여 판매 물건 출력

# 이진탐색 클래스 정의
class TreeNode:
    def __init__(self): 
        self.left = None 
        self.data = None 
        self.right = None

# 트리 루트 노드 생성
root = TreeNode() 
root.data = sellAry[0]  # 첫 번째 데이터를 root로
memory.append(root) # 메모리리스트에 저장


for name in sellAry[1:]: 
    node = TreeNode()
    node.data = name

    current = root
    while True:
        if name == current.data:  # 중복된 경우 삽입하지 않음
            break
        if name < current.data:  # 왼쪽 서브트리로 이동
            if current.left is None:
                current.left = node
                memory.append(node)
                break
            current = current.left
        else:  # 오른쪽 서브트리로 이동
            if current.right is None:
                current.right = node
                memory.append(node)
                break
            current = current.right

print('이진 탐색 트리 구성 완료!')

# **전위 순회 함수 **
def preorder(node):
    if node is None:
        return
    print(node.data, end=' ')  # 현재 노드 출력
    preorder(node.left)  # 왼쪽 서브트리 탐색
    preorder(node.right)  # 오른쪽 서브트리 탐색

print('오늘 판매된 종류(중복X) -->', end=' ')
preorder(root)
print()
```

4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
import random

# 변수 선언, 이진 검색시에는 순차적으로 정렬 되어있어야한다. 

findData = random.randint(0, 99999) # 랜덤하게 0~99999 사이의 숫자 하나를 선택 (찾을 숫자)
dataAry = [random.randint(0, 99999) for _ in range(100000)] # 0~99999 사이의 랜덤한 숫자 100000개를 리스트에 생성
dataAry.insert(random.randint(0, 100000), findData)# 리스트의 임의 위치에 찾을 숫자 삽입


sortedAry = sorted(dataAry) # 이진 검색을 위해 리스트 정렬

count = 0 # 처리 횟수 카운트 변수

def BinSearch(ary, data):
    global count # 처리횟수를 알기위해 count = 0 초기화 후 Global 사용
    pos = -1 # 찾지 못했을 때 반환할 값
    start = 0 # 배열의 시작점
    end = len(ary) - 1 # 리스트 마지막 인덱스

    while (start <= end): # 시작이 끝보다 커지면 검색종료
        mid = (start + end) // 2  #  중간 값 계산
        if data == ary[mid]: # 값을 찾은경우
            count += 1  # 처리횟수를 알기위해 넣어봄
            return mid # 찾은 위치 반환 mid 가 다시 while 상단으로 가 중간값 계산이 됨
        elif data > ary[mid]: # 값이 찾는값보다 큰 경우
            start = mid + 1 # 오른쪽을 탐색 ( +1 )           
        else:
            end = mid - 1 # 찾는값보다 작은 경우는 왼쪽탐색 ( -1 )       
        count += 1 # 처리횟수를 알기위해 넣어봄

    return pos # pos 반환 못찾았을 경우 -1 로 반환되어 data 없음을 출력 (하단 if 문)


pos = BinSearch(sortedAry,findData)
if pos == -1: # 못찾은경우
    print(f'{findData} 없음')  # 없음출력

else: # 그렇지않다면
    print(f'{findData} / {pos + 1}번째 자리에 존재') # 문자 포멧팅한 입력한 찾을값, pos+1로 자릿수출력 
# 정렬된 배열 확인 (앞 5개와 뒤 5개 출력)
print('# 정렬된 배열(10만개) --> ', sortedAry[:5], '~~~', sortedAry[-5:])
print(f'총 처리 횟수: {count}')


```
